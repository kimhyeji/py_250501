```python
fig = plt.figure()

a1 = fig.add_subplot(1, 2, 1)
a2 = fig.add_subplot(1, 2, 2)

x = range(0, 10)
y = range(5, 15)
a1.plot(x, y, "y^")

x = range(0, 50)
y = range(0, 50)
a2.plot(x, y, "k.")

plt.show()
```

```python
import matplotlib
import matplotlib.font_manager as fm

# 간격을 가능한 최적의 수치로 자동 조정
fig = plt.figure(constrained_layout=True)

# 원하는 폰트 파일 경로 문자열로 입력
font_location = "C:\\Windows\\Fonts\\HMFMPYUN.TTF"

# 지정한 폰트 파일 경로에서 폰트 추출
fontname = fm.FontProperties(fname=font_location).get_name()

# matplotlib의 전역 폰트 설정 변경
matplotlib.rc('font', family=fontname)

a1 = fig.add_subplot(1, 2, 1)
a2 = fig.add_subplot(1, 2, 2)

x = range(0, 10)
y = range(5, 15)
a1.plot(x, y, "y^")

x = range(0, 50)
y = range(0, 50)
a2.plot(x, y, "k.")

# x축, y축 이름 지정
a1.set_xlabel('시간')
a1.set_ylabel('돈')

a2.set_xlabel('시간')
a2.set_ylabel('돈')

plt.show()
```

```python
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] # 막대의 위치(좌표 기준)
y = [1, 2] # 막대의 높이(값 기준)

# x축 라벨 이름(tick 이름)
labels = ['english', 'math']

# 각 막대에 적용할 색상
colors = ['red', 'blue']

a.bar(x, y, tick_label=labels, color=colors)
plt.show()
```

```python
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] # 막대의 위치(좌표 기준)
y = [1, 2] # 막대의 높이(값 기준)

# x축 라벨 이름(tick 이름)
labels = ['english', 'math']

# 각 막대에 적용할 색상
colors = ['red', 'blue']

a.barh(x, y, tick_label=labels, color=colors) # h를 붙이면 가로로 나옴
plt.show()
```

```python
# 막대 그래프에 수치 표현
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] # 막대의 위치(좌표 기준)
y = [1, 2] # 막대의 높이(값 기준)

# x축 라벨 이름(tick 이름)
labels = ['english', 'math']

# 각 막대에 적용할 색상
colors = ['red', 'blue']

a.bar(x, y, tick_label=labels, color=colors)

for i, v in enumerate(y):
  a.text(v+2, i + 1.05, str(v)) # 표준 수치

plt.show()
```
